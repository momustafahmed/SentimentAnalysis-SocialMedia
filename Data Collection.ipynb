```python
# The video URL: https://youtu.be/APfKrSnF20o

youtube = build('youtube', 'v3', developerKey=API_KEY)

def get_video_comments(video_id):
    comments = []
    response = youtube.commentThreads().list(
        part='snippet,replies',
        videoId=video_id,
        maxResults=2150,  # Adjust based on your needs
        textFormat='plainText'
    ).execute()

    while response:
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments.append({
                'author': comment['authorDisplayName'],
                'date': comment['publishedAt'],
                'Comment': comment['textDisplay'],
                'like_count': comment['likeCount'],
                'reply_count': item['snippet']['totalReplyCount'],
                'comment_length': len(comment['textDisplay']) 
            })

            # If there are replies, fetch them as well
            if item['snippet']['totalReplyCount'] > 0:
                for reply in item['replies']['comments']:
                    reply_comment = reply['snippet']
                    comments.append({
                        'author': reply_comment['authorDisplayName'],
                        'date': reply_comment['publishedAt'],
                        'Comment': reply_comment['textDisplay'],
                        'like_count': reply_comment['likeCount'],
                        'reply_count': 0, # Replies do not have further replies
                        'comment_length': len(reply_comment['textDisplay']) 
                    })

        # Check if there are more pages of comments
        if 'nextPageToken' in response:
            response = youtube.commentThreads().list(
                part='snippet,replies',
                videoId=video_id,
                pageToken=response['nextPageToken'],
                maxResults=100,  # Adjust based on your needs
                textFormat='plainText'
            ).execute()
        else:
            break

    return comments

# Example usage
video_id = 'APfKrSnF20o'
comments = get_video_comments(video_id)
for comment in comments:
    print(comment)
```

### Merge all the videos

```python
import pandas as pd
import os

# List of CSV file paths

csv_files = [
    'Qoslaaye1.csv',
    'Qoslaaye2.csv',
    'Qoslaaye3.csv',
    'Qoslaaye4.csv',
    'Qoslaaye5.csv',
    'Qoslaaye6.csv',
    'Qoslaaye7.csv',
    'Qoslaaye8.csv',
    'Qoslaaye9.csv',
    'Qoslaaye10.csv',
]

# Initialize an empty list to store DataFrames
dfs = []

# Loop through the list of CSV files
for file in csv_files:
    # Check if the file exists to avoid errors
    if os.path.exists(file):
        # Read the CSV file into a DataFrame and append it to the list
        Data = pd.read_csv(file)
        dfs.append(Data)
    else:
        print(f"File not found: {file}")

# Concatenate all DataFrames in the list into one
combined_df = pd.concat(dfs, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('Qoslaaye_All.csv', index=False)

print("All CSV files have been combined into one.")
```
